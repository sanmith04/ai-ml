```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('F:\image processing\gradient.jpg')
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4=cv2.threshold(img,122,255,cv2.THRESH_TOZERO)
ret,thresh5=cv2.threshold(img,122,255,cv2.THRESH_TOZERO_INV)

titles=['ORIGNAL Image','BInary image','binary_inv image','trunc image','tozero img','tozero_inv image']
images=[img,thresh1,thresh2,thresh3,thresh4,thresh5]


for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show
```

```python
img=cv2.imread('F:\image processing\morpho.jpg',0)
kernal=np.ones((5,5),np.uint8)
erosn=cv2.erode(img,kernal,iterations=1)
plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')


plt.subplot(122),plt.imshow(erosn),plt.title('Erode Image')

plt.show()
```

```python
dilate=cv2.dilate(img,kernal,iterations=1)

plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')

plt.subplot(122),plt.imshow(dilate),plt.title('dilate Image')
```

```python
img=cv2.imread('F:\image processing\download.jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(img_gray,127,255,0)

contours,hierarchy=cv2.findContours(thresh,2,1)
cnt=contours[0]

hull=cv2.convexHull(cnt,returnPoints=False)
defects=cv2.convexityDefects(cnt,hull)

for i in range(defects.shape[0]):
    s,e,f,d=defects[i,0]
    start=tuple(cnt[s][0])
    end=tuple(cnt[e][0])
    far=tuple(cnt[f][0])
    cv2.line(img,start,end,[0,255,0],2)
    cv2.circle(img,far,5,[0,0,255],-1)
    
cv2.imshow('CONTOUR_Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
