```python
img=cv2.imread('F:\image processing\download.jpg',0)

rows,col=img.shape
```

```python
img.shape
```

```python
x=np.float32([[1,0,100],[0,1,50]])
trans=cv2.warpAffine(img,x,(col,rows))
cv2.imshow("WArpaffine",trans)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
X=cv2.getRotationMatrix2D((col/2,rows/2),60,1)
rotate=cv2.warpAffine(img,X,(col,rows))
re_size=cv2.resize(rotate,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
cv2.imshow("rotated image",re_size)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.warpPerspective

```python
import matplotlib.pyplot as plt
img=cv2.imread('F:\image processing\download.jpg')
rows,cols,ch=img.shape
```

```python
pos1=np.float32([[56,65],[300,87],[16,387],[289,379]])
pos2=np.float32([[0,0],[300,0],[0,300],[300,300]])

x=cv2.getPerspectiveTransform(pos1,pos2)

rotate_rgb_img=cv2.warpPerspective(img,x,(200,200))

plt.subplot(121),plt.imshow(img),plt.title("ORGINAL PIC")
plt.subplot(122),plt.imshow(rotate_rgb_img),plt.title("WARPERSPEVTIVE IMG")
```

# Smoothing

```python
kernal=np.ones((5,5),np.float32)/30
```

```python
kernal
```

```python
blur_img=cv2.filter2D(img,-1,kernal)

plt.subplot(121),plt.imshow(img),plt.title('Actual Input Image')

plt.subplot(122),plt.imshow(blur_img),plt.title('Averaging blur image')

plt.show()
```

# GAUSSian blur

```python
gauss_blur=cv2.GaussianBlur(img,(5,5),-1)
plt.subplot(121),plt.imshow(img),plt.title("actual img")

plt.subplot(122),plt.imshow(gauss_blur),plt.title("blur img")
plt.show()
```
